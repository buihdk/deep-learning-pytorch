```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)

def drawGraph(m, b, color='g--'):
    plt.xlim(-0.05,1.05)
    plt.ylim(-0.05,1.05)
    x = np.arange(-10, 10, 0.1)
    plt.plot(x, m*x+b, color)

def stepFunction(t):
    if t >= 0:
        return 1
    return 0

def prediction(X, W, b):
    return stepFunction((np.matmul(X,W)+b)[0])

def perceptronStep(X, y, W, b, learn_rate = 0.01):
    # Fill in code
    for i in range(len(X)):
        y_hat = prediction(X[i], W, b)
        if y[i] - y_hat == 1: # missclassified as positive
            W[0] += X[i][0]*learn_rate
            W[1] += X[i][1]*learn_rate
            b += learn_rate
        elif y[i] - y_hat == -1: # missclassified as negative
            W[0] -= X[i][0]*learn_rate
            W[1] -= X[i][1]*learn_rate
            b -= learn_rate
    return W, b

def trainPerceptronAlgorithm(X, y, learn_rate = 0.01, num_epochs = 30):
    x_min, x_max = min(X.T[0]), max(X.T[0])
    y_min, y_max = min(X.T[1]), max(X.T[1])
    W = np.array(np.random.rand(2,1))
    b = np.random.rand(1)[0] + x_max
    # These are the solution lines that get plotted below.
    boundary_lines = []
    for i in range(num_epochs):
        # In each epoch, we apply the perceptron step.
        W, b = perceptronStep(X, y, W, b, learn_rate)
        boundary_lines.append([-W[0]/W[1], -b/W[1], b])
        drawGraph(-W[0]/W[1], -b/W[1])
    drawGraph(-W[0]/W[1], -b/W[1], 'black')
    return boundary_lines
```

```python
data = pd.read_csv("data.csv") 
x1 = data['X1'].tolist()
x2 = data['X2'].tolist()
plt.plot(x1[:50], x2[:50], 'ro', x1[50:], x2[50:], 'bo')
plt.show()
```

```python
X = []
for i in range(len(x1)):
    X.append([x1[i], x2[i]])
X = np.array(X)

y = data['Y'].tolist()
y = np.array(y)
trainPerceptronAlgorithm(X, y)

plt.plot(x1[:50], x2[:50], 'ro', x1[50:], x2[50:], 'bo')
plt.title("Perceptron Learning Algorithm")
plt.show()
```
